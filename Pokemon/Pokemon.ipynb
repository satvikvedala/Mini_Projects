```python
from ipywidgets import widgets, interactive
name = widgets.Dropdown(
    options = list(Gen1['name']),
    description = 'Pokemon name'
)
def plotit(name):
    fig,ax = plt.subplots(figsize = (10,8))
    y = ['attack','defense','sp_attack','hp','sp_defense','speed']
    x = [int(Gen1.loc[Gen1['name'] == name]['attack']),int(Gen1.loc[Gen1['name'] == name]['defense']),
        int(Gen1.loc[Gen1['name'] == name]['sp_attack']),int(Gen1.loc[Gen1['name'] == name]['hp']),
        int(Gen1.loc[Gen1['name'] == name]['sp_defense']),int(Gen1.loc[Gen1['name'] == name]['speed'])]
    txtstr = 'name: {name}\nType:{type1}\nGeneration: 1'.format(name = name,type1 = Gen1.loc[Gen1['name'] == name]['type1'][0])
    ax.text(0.1, 6, txtstr, fontsize=14)
    ax.barh(y,x)
    ax.tick_params(axis='both', labelsize=14)
    for item in ['top','bottom','left','right']:
        ax.spines[item].set_visible(False)
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticklabels([])
    ax.yaxis.set_tick_params(pad = 10)
    for i in ax.patches:
        ax.text(i.get_width()+0.2,i.get_y()+0.4,str(round((i.get_width()),2)),fontsize = 14)
    plt.show()
interactive(plotit,name = name)
```

```python
name = widgets.Dropdown(options = list(Gen1['name']),description = 'Pokemon name')
def plotit(name):
    categories = ['speed','sp_defense','hp','sp_attack','defense','attack']
    values = [int(Gen1.loc[Gen1['name'] == name]['speed'])/max(Gen1['speed']),
              int(Gen1.loc[Gen1['name'] == name]['sp_defense'])/max(Gen1['sp_defense']),
              int(Gen1.loc[Gen1['name'] == name]['hp'])/max(Gen1['hp']),
              int(Gen1.loc[Gen1['name'] == name]['sp_attack'])/max(Gen1['sp_attack']),
              int(Gen1.loc[Gen1['name'] == name]['defense'])/max(Gen1['defense']),
              int(Gen1.loc[Gen1['name'] == name]['attack'])/max(Gen1['attack'])]
    label_loc = np.linspace(start = 0,stop = 2*np.pi,num = len(values))
    ax = plt.figure(figsize = (10,8))
    plt.subplot(polar = True)
    plt.plot(label_loc,values)
    lines,labels = plt.thetagrids(np.degrees(label_loc),labels = categories)
    plt.show()
interactive(plotit,name = name)
```
